```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('datasets/covid19.csv')
```

```python
df
```

```python
#Kita mau bandingkan dan analisis negara indonesia dan singapura, jumlah kasus terkonfirmasi covid 2020
```

```python
#kita lakukan filtering indonsia
data_indo = df[df['Country/Region'] == 'Indonesia']
```

```python
data_indo
```

```python
# kita filtering juga singapura
data_singapura = df[df['Country/Region']== 'Singapore']
```

```python
data_singapura
```

```python
#kita filter untuk data yang terkonfimasi covid saja
conf_indo = data_indo[data_indo['Confirmed']>0]

conf_singa = data_singapura[data_singapura['Confirmed']>0]
```

```python
conf_indo
```

```python
conf_singa
```

```python
#menyamakan sumbu x berdasarkan hari, ubah time data date
```

```python
conf_indo['Date'] = pd.to_datetime(conf_indo['Date'])

conf_singa['Date'] = pd.to_datetime(conf_singa['Date'])
```

```python
conf_indo.head()
```

```python
conf_singa.head()
```

```python
#buat kolom basu untuk menyamakan date untuk tau hari ke berapa
conf_indo['Basedate'] = pd.to_datetime('2020-03-02')
conf_singa['Basedate'] = pd.to_datetime('2020-01-23')

```

```python
conf_indo.head()
```

```python
#buat kolom Hari ke
conf_indo['Harike'] = conf_indo['Date'] - conf_indo['Basedate']

conf_singa['Harike'] = conf_singa['Date'] - conf_singa['Basedate']
```

```python
conf_singa.head()
```

```python
#ubah hari ke numerik biar lebih mudah diolah

conf_indo.Harike = conf_indo.Harike/pd.Timedelta(1, unit='d')

conf_singa.Harike = conf_singa.Harike/pd.Timedelta(1, unit='d')
```

```python
#visualisasi kan ke grafik

plt.plot(conf_indo.Harike, conf_indo.Confirmed, color='red',ls='-.', marker='+', markersize=1)
plt.plot(conf_singa.Harike, conf_singa.Confirmed, ls = ':')
plt.title('junmla perbandingan kasus covid indonesia dan singapura')
plt.xlabel('Hari ke-')
plt.ylabel('Jumlah confirmed corona')
plt.legend(['Indonesia', ['Singapore']])
plt.xlim(0,85)
plt.show()
```
